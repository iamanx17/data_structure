```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
class Queue:
    def __init__(self):
        self.__head=None
        self.__tail=None
        self.__count=0
    
    def enqueue(self,data):
        newnode=Node(data)
        if self.__head is None:
            self.__head=newnode
        else:
            self.__tail.next=newnode
        self.__tail=newnode
        self.__count+=1
    
    def dequeue(self):
        if self.__head is None:
            return 0
        data=self.__head.data
        self.__head=self.__head.next
        self.__count-=1
        return data
    
    def front(self):
        if self.__head is None:
            return 0
        data = self.__head.data
        return data
    def size(self):
        return self.__count
    
    def isEmpty(self):
        return self.size()==0
```

```python
class Tree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

def printdata(root):
    if root is None:
        return
    print(root.data,end=':')
    if root.left is not None:
        print('L',root.left.data, end=',')
    if root.right is not None:
        print('R', root.right.data)
    print()
    
    printdata(root.left)
    printdata(root.right)


def takeinput():
    q=Queue()
    print('Enter root')    
    data=int(input())
    
    if data == -1:
        return None
    
    root=Tree(data)
    q.enqueue(root)
    
    while not q.isEmpty():
        curr=q.front()
        print(curr)
        
        print("Enter left child of", curr.data)
        leftchild = int(input())
        if leftchild!=-1:
            leftchild=Tree(leftchild)
            curr.left=leftchild
            q.enqueue(leftchild)
            
        print('Enter right child of', curr.data)
        rightchild = int(input())
        if rightchild != -1:
            rightchild = Tree(rightchild)
            curr.right=rightchild
            q.enqueue(rightchild)
    
    return root
    

    
    
```
