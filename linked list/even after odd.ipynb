```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None    

def takeinput():
    inputlist=[int(ele) for ele in input().split()]
    head=None
    tail=None
    for currdata in inputlist:
        if currdata==-1:
            break    
            
        newnode=Node(currdata)
        
        if head is None:
            head=newnode
            tail=newnode
        else:
            tail.next=newnode
            tail=newnode
            
    return head

def printdata(head):
    while head is not None:
        print(str(head.data), end='->')
        head=head.next
    print('None') 


def mergesort(head1,head2):
    head=None
    tail=None
    if head1.data<head2.data:
        head=head1
        tail=head1
        head1=head1.next
    else:
        head=head2
        tail=head2
        head2=head2.next        
    while head1.next !=None and head2.next !=None:
        if head1.data<=head2.data:
            tail.next=head1
            tail=head1
            head1=head1.next
        else:
            tail.next=head2
            tail=head2
            head2=head2.next
    if head1.next is None:
        tail.next=head2
    if head2.next is None:
        tail.next=head1
    return head

def sort(head):
    middle=mid(head)
    nextofmid=middle.next
    middle.next=None
    reverse=mergesort(head, nextofmid)
    return reverse
def mid(head):
    slow=head
    fast=head
    while fast.next is not None and fast.next.next is not None:
        slow=slow.next
        fast=fast.next.next
    return slow

def evenfodd(head):
    even=None
    odd=None
    while head.next is not None:
        nxt=head.next        
        if head.data %2==0:            
            head.next=even
            even=head
            head=nxt
        else:
            head.next=odd
            odd=head
            head=nxt    
    if head.next is None and head.data%2==0:
        head.next=even
        even=head
    else:
        head.next=odd
        odd=head
    
    neweven=sort(even)
    newodd=sort(odd)
    source=newnodd
    while odd.next is not None:
        odd=odd.next
    odd.next=even
    return source
    
        
```
