```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None    

def takeinput():
    inputlist=[int(ele) for ele in input().split()]
    head=None
    tail=None
    for currdata in inputlist:
        if currdata==-1:
            break
        
        newnode=Node(currdata)
        
        if head is None:
            head=newnode
            tail=newnode
        else:
            tail.next=newnode
            tail=newnode
    return head

def printdata(head):
    while head is not None:
        print(str(head.data), end='->')
        head=head.next
    print('None')     

def swap(head,i,j):
    if i-j==1 or j-i==1: 
        source=head
        count=0
        while count<i:
            curr=head
            head=head.next
            count+=1
        save=head.next.next
        second=head.next
        curr.next=second
        second.next=head
        head.next=save
        return source
    else:  
        source=head
        count=0
        while count<i:
            curr=head
            head=head.next
            count+=1   
        first=head
        head=source
        count=0
        while count<j:
            acurr=head
            head=head.next
            count+=1        
        nxt1=first.next
        nxt2=head.next    
        curr.next=head
        head.next=nxt1       
        acurr.next=first
        first.next=nxt2    
        return source           
            
```

```python
head=takeinput()
printdata(head)
m=int(input())
n=int(input())
printdata(swap(head,m,n))
```
